```python
import time
from math import sin, cos, sqrt, atan2, radians, pow
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
a=pd.read_csv(r"C:\\Users\Ogeid99\Desktop\cataloguess\DATA3_magn.csv")
b=pd.read_csv(r"C:\\Users\Ogeid99\Desktop\cataloguess\DATA_reasenberg.csv")
c=pd.read_csv(r"C:\\Users\Ogeid99\Desktop\cataloguess\DATA_correlation_metrics_20000_3600.csv")
```

```python
#fig.suptitle('HISTOGRAMS',fontsize=32)
fig,axes=plt.subplots(1,1,figsize=(10,6))
a['Time'].astype("datetime64")
b['Time'].astype('datetime64')
c['Time'].astype('datetime64')
from datetime import datetime
#convert datetime 64 to datetime
dt = datetime.utcnow() 
dt64 = np.datetime64(dt)
time_new1=[]
time_new2=[]
time_new3=[]
for i in a['Time']:
    ts = (dt64 - np.datetime64(i)) / np.timedelta64(1, 's')
    times=datetime.utcfromtimestamp(ts)
    time_new1.append(times)
a['Time']=time_new1
for i in b['Time']:
    ts = (dt64 - np.datetime64(i)) / np.timedelta64(1, 's')
    times=datetime.utcfromtimestamp(ts)
    time_new2.append(times)
b['Time']=time_new2
for i in c['Time']:
    ts = (dt64 - np.datetime64(i)) / np.timedelta64(1, 's')
    times=datetime.utcfromtimestamp(ts)
    time_new3.append(times)
c['Time']=time_new3
```

```python
fig,axes=plt.subplots(1,1,figsize=(10,6))
import matplotlib.patches as mpatches
plt.hist([a['Time'],b['Time'],c['Time']],bins=60,color=['aquamarine','green','black'],cumulative=True)#,color='coral')
#plt.hist(b['Time'],bins=90,color='coral')
#plt.hist(c['Time'],bins=90,color='coral')
plt.xlabel("Time",fontsize=20) #set font and title of x label
plt.ylabel("Frequency",fontsize=20) #set font and title of y label     #plt instead of set_
#plt.set_yscale('log') #set log scale
plt.tick_params(labelsize=14) #change labelsize
plt.title('Number of earthquakes $\geq$ 3 throughout the years (cumulative)',fontsize=14,pad=5.8)
aquamarine_patch = mpatches.Patch(color='aquamarine', label='No declustering')
green_patch = mpatches.Patch(color='green', label='Reasenberg')
black_patch = mpatches.Patch(color='black', label='Correlation Metric')
plt.legend(handles=[black_patch,green_patch,aquamarine_patch],loc="upper left")
#you can speak about Collefiorito, L'Aquila, Amatrice and completeness through the years
fig.savefig(r"C:\\Users\Ogeid99\Desktop\tesi\photos_thesis\grafici_finali_jupyter\frequency_time_hist_declust_total_cumulative.png",dpi=600)
```
