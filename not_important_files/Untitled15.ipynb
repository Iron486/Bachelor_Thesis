```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
b=DATA.sort_values(by='Time')
counter=pd.Series([1 for i in range(len(DATA['Time']))])
counter.name='Count'
b=pd.concat([b,pd.Series(counter)],axis=1)
b=b['Count'].groupby(b['Time'])
b_sum=pd.Series(b.sum())
#sns.distplot(b_sum,kde=False) # frequency of earthquakes per count
#b_sum.index().tolist(b_sum)
b_sum_index=b_sum.index.values
plt.bar(b_sum_index,b_sum,edgecolor='navy') #edgecolor: bar color
plt.xlabel('Time')
plt.ylabel('N. of earthquakes')
```
