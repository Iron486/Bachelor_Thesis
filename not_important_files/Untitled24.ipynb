```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings

a=pd.read_csv(r"C:\\Users\Ogeid99\Desktop\cataloguess\DATA_no_modify.csv")
a
```

```python
import matplotlib as mpl
norm = plt.Normalize(0, 5000)
#color = cmap(norm(100.))
bins = np.logspace(np.log10(1),3,num=40)
plt.hist2d(a['Magnitude'],a['Depth/Km'],bins=(43,bins),cmap=plt.cm.RdBu_r,norm=norm)#,norm=mpl.colors.LogNorm())
plt.yscale('log')
plt.colorbar()
plt.ylim(2,200)
plt.xlim(1.5,4)
```

```python
condition_a=a[ (a['Depth/Km']>0.1 )&(a['Longitude']<17.01 ) &(a['Longitude']>10.01)&(a['Latitude']<41.01)&(a['Latitude']>38.01) ]
condition_a
condition_a['Boolean']=[1 for i in range(condition_a['Magnitude'].size)]
condition_a
```

```python
a=pd.concat([a,condition_a['Boolean']],axis=1)
a['Boolean']=a['Boolean'].fillna('0')
a
```

```python
fig,axes=plt.subplots(1,1,figsize=(10,6))

sns.scatterplot(a['Magnitude'],a['Depth/Km'],hue=a['Boolean'])
plt.ylim(0.07,1080)
plt.yscale('log')
plt.ylabel('Depth [km]',fontsize=17,labelpad=5)
plt.xlabel('Magnitude',fontsize=17,labelpad=5)
#plt.title('Number of earthquakes greater than 1.5 by depth',fontsize=18,pad=15.0)  
plt.tick_params(labelsize=12)
plt.legend().set_visible(False)
plt.savefig(r"C:\\Users\Ogeid99\Desktop\tesi\photos_thesis\grafici_finali_jupyter\magnitude_depth_graph_hue.png",dpi=600)
```
