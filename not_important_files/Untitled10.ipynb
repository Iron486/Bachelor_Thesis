```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
a=pd.read_csv("C:\\Users\Ogeid99\Desktop\DS\excel\catalogo_l'aquila.txt",sep='|')
a.sort_values(by='Time',ascending=True) #sort values according to time
num=pd.Series([i for i in range(int(a.size/len(a.columns)))])
a=pd.concat([a,num],axis=1)
#a.set_index(0)
#a.index=[i for i in range(int(a.size/len(a.columns)))]
a
```

```python
from datetime import datetime
from math import sin, cos, sqrt, atan2, radians, pow
d=1.6   #it could change... try different values
b=1.5
a_time=pd.Series(a['Time'])
a['Time']=a_time.astype(np.datetime64)
#print(a['Time'])
for j in range(int(a.size/len(a.columns))-9100):  #a.size: size of data frame, a.columns: size of the columns
    #access the first (only) row as a Series using iloc, and then the value using the column name
    timej=a_time.iloc[j]#['Time']   
    latj = radians(a.iloc[j]['Latitude'])
    lonj = radians(a.iloc[j]['Longitude'])
    R = 6373.0    # approximate radius of earth in km
    i=1
    space_time_list=[]
    while i+j <= int(a.size/len(a.columns))-9100 : 
        timei=a_time.iloc[i+j]#['Time']
        dtime=((timei-timej)).total_seconds()
        magnitude=a.iloc[i+j]['Magnitude']
        lati = radians(a.iloc[i+j]['Latitude'])
        loni = radians(a.iloc[i+j]['Longitude'])
        dlon = loni - lonj
        dlat = lati - latj
        a2 = sin(dlat / 2)**2 + cos(latj) * cos(lati) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a2), sqrt(1 - a2))
        distance = R * c
        space_time=dtime*distance**d*pow(10,-b*float(magnitude))
        space_time_list.append(space_time)
        i+=1
    print(space_time_list)
        
```
