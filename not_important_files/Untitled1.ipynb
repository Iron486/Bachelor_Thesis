```python
list_condition=[]
list_years=[]
for i in range(0,36,2):
    tc=DATA[(DATA['Time']>=datetime(1985+i,1,1,0,0,0))& (DATA['Time']<=datetime(1985+i,12,31,23,59,59) )]
    list_condition.append(tc)
    list_years.append(1985+i)
#fig=plt.figure()
fig,axes=plt.subplots(6,3,figsize=(20,24))
sns.set(style="ticks", palette="pastel")
fig.suptitle('HISTOGRAMS',fontsize=32)
# Draw a nested boxplot 
k=0
j=0
list_complete=[]
for i,year in zip(list_condition,list_years):
    if j==0:      
        axes[k][j].hist(i['Magnitude'])
        list_complete.append([year,k,j])
        if k==5:
            j=1
            k=-1
    elif j==1:
        axes[k][j].hist(i['Magnitude']) 
        list_complete.append([year,k,j])
        if k==5:
            j=2
            k=-1 
    elif j==2:
        axes[k][j].hist(i['Magnitude']) 
        list_complete.append([year,k,j])
    k+=1
for i in list_complete:
    ax = axes[i[1],i[2]]
    if i[2]>=1:
        title = ax.set_xlabel(str(i[0]),fontsize=25) #set font and title of x label
        title2 = ax.set_ylabel("",fontsize=22) #set font and title of y label
        ax.set_yscale('log') #set log scale    
        ax.tick_params(labelsize=16) #change labelsize 
    else:
        title = ax.set_xlabel(str(i[0]),fontsize=25) #set font and title of x label
        title2 = ax.set_ylabel("Frequency",fontsize=22) #set font and title of y label
        ax.set_yscale('log') #set log scale
        ax.tick_params(labelsize=16) #change labelsize 
        
#ax1.title.set_text('First Plot') title of the first plot
#comment this histograms
plt.subplots_adjust(wspace = 0.2, hspace=0.8) #spaces among plots
plt.savefig('figcompleteness_time.pdf',bbox_inches='tight')
```
