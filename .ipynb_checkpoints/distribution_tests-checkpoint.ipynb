```python
import seaborn as sns
import pandas as pd
import numpy as np
x = np.random.normal(size=1000)
sns.distplot(x,bins=70,kde=False,rug=True,color='g') #kd=False cancel the line. #rug=True adds small vertical thicks. #bins:number of
#'columns' in the histogram
```

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 1000)
df = pd.DataFrame(data, columns=["x", "y"])
df
#jointplot() function creates a multi-panel figure that shows 
#both the bivariate (or joint) relationship between two variables
#along with the univariate (or marginal) distribution of each on separate axes.
sns.jointplot(x="x", y="y", data=df, kind='hex', color='b') #kind hex plots hezagons instead of points
```
