```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
# Dataset
DATA=pd.read_csv(r"C:\\Users\Ogeid99\Desktop\cataloguess\DATA_no_modify.csv")
# plot
DATA['Time'] = pd.to_datetime(DATA['Time'])
lat=(DATA['Latitude']<42.7) & (DATA['Latitude']>42.2)
long=(DATA['Longitude']<13.72)& (DATA['Longitude']>13.09) 
depth=(DATA['Depth/Km']<1000)& (DATA['Depth/Km']>0) 
magn=(DATA['Magnitude']>2.1)
tm=(DATA['Time']< datetime(2010,1,3,0,0,0))& (DATA['Time']>=datetime(2008,10,31,0,0,0) ) #modify interval of time
DATA_Aquila= DATA[tm&lat&long&magn]
fig = plt.figure()
ax = fig.add_subplot(111) #, projection='3d')
#colour=excel_data[1]
sns.scatterplot(DATA_Aquila['Depth/Km'],DATA_Aquila['Latitude'],s=9)
#plt.gca().invert_zaxis() #it inverts z axis
#plt.rcParams['xtick.labeltop']
#ax.set_title('Eartquakes in 3D')

#ax.show()
```

```python
 #change conditions
condition=DATA['Magnitude']>2.9
latitude=(DATA['Latitude']<43.2) &(DATA['Latitude']>42.3)
longitude=(DATA['Longitude']>12.8) &(DATA['Longitude']<13.4)
```

```python
DATA_windowed=DATA[(condition) & (latitude )& (longitude)]
```
