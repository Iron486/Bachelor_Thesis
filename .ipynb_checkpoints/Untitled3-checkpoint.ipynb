```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
df=pd.read_csv('C:\\Users\Ogeid99\Desktop\TABLE(declust).txt', sep= '\t', header=None) #copy a table from  matlab to text
#and then read the csv 
df.columns=np.array(['Date', 'Magnitude' ,'MagnitudeType', 'DecimalYear', 'DayOfYear',
                            'X' ,'Y', 'Z', 'Longitude', 'Latitude', 'Depth'],dtype=object)
from datetime import datetime
df_slice=df['Date']
df_notimportant=df['MagnitudeType']
function=lambda x: x.replace("'","") #replace "'" with ' '
datetime_object = lambda x: datetime.strptime(x, '%d-%b-%Y %H:%M:%S')
df_not_important=df_notimportant.map(function)
df_slice=df_slice.map(function)
df_slice=df_slice.map(datetime_object)
df_slice=df_slice.astype(np.datetime64)
timedelta=[]
df_slice_numpy=df_slice.to_numpy()
np.set_printoptions(suppress=True) #it's revealed useless
m=0
for i in df_slice_numpy:
    m+=1
    if m>0 and m < len(df_slice_numpy): 
        #the second condition is written to avoid the error due to exceedance of array lenght  
        diff=(df_slice_numpy[m]-df_slice_numpy[m-1])/(1000000000*60)
        diff=diff.astype(np.int32)#convert in int
        timedelta.append(int(diff))
#lis_timedelta=[int(n) for n in timedelta] #procedure to cut the zero out of the number in the array,
#creating a list and then retransform in an array
timedeltas  =np.array(timedelta)
timedeltas
```
